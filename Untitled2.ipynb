# Модель Лотки-Вольтерра: Стартапи vs Монополісти



Ця модель описує динаміку взаємодії між стартапами та монополістами на ринку.

За аналогією з класичною моделлю "хижак-жертва":
Стартапи (x) - "жертви": зростають самостійно, але зникають під тиском монополістів
Монополісти (y) - "хижаки": розвиваються за рахунок поглинання стартапів

Параметри моделі:
alpha - природний темп зростання стартапів;
beta - інтенсивність поглинання стартапів монополістами;
delta - ефективність, з якою монополісти "засвоюють" стартапи;
gamma - природний темп зникнення монополістів без нових поглинань

```python
import numpy as np
x, y, t = var('x y t')
alpha, beta, delta, gamma = 0.5, 0.02, 0.01, 0.3
f1 = alpha*x - beta*y
f2 = delta*x*y - gamma*y
points = solve([f1 == 0, f2 == 0], x, y)
print(f"Точки рівноваги: {points}")
v_field = plot_vector_field((f1, f2), (x, 0, 100), (y, 0, 1500), color='pink', title="Фазовий портрет: Стартапи vs Монополісти")
t_span = np.linspace(0, 50, 500)
ics = [20, 30]
sol = desolve_odeint([f1, f2], ics, t_span, [x, y])
trajectory = line(sol, color='red', thickness=2)
p_time = list_plot(list(zip(t_span, sol[:,0])), plotjoined=True, color='green', legend_label='Стартапи')
p_time += list_plot(list(zip(t_span, sol[:,1])), plotjoined=True, color='red', legend_label='Монополісти')
p_time.axes_labels(['Час', 'Кількість'])
(v_field + trajectory).show(figsize=6, axes_labels=['Стартапи', 'Монополісти'])
p_time.show(figsize=6)
                            
```

| Параметр | Значення | Опис |
|----|----|----|
| α (alpha) | 0.5 | Природний темп зростання стартапів |
| β (beta) | 0.02 | Інтенсивність поглинання стартапів монополістами |
| δ (delta) | 0.01 | Ефективність засвоєння стартапів монополістами |
| γ (gamma) | 0.3 | Природний темп зникнення монополістів |

Інтерпретація точок рівноваги

Модель має дві точки рівноваги:

1. (0, 0) - тривіальна: обидві популяції вимерли. 
   Ринок порожній - немає ні стартапів, ні монополістів.

2. (30, 750) - нетривіальна: стійкий стан ринку.
   При 30 стартапах і 750 монополістах система врівноважена - 
   поглинання компенсується природним зростанням нових гравців.

 Висновок

Модель Лотки-Вольтерра демонструє, що ринок стартапів і монополістів 
поводиться циклічно: стартапи зростають, монополісти поглинають їх і 
теж зростають, нестача стартапів призводить до занепаду монополістів, 
стартапи знову зростають.

Система має дві точки рівноваги:
 *(0, 0)* - ринок повністю вимер
 *(30, 750)* - стійкий стан співіснування

Реальна поведінка системи залежить від початкових умов: при певних 
значеннях траєкторія може виходити за межі стійкої зони, що відповідає 
ринковим кризам або монополізації.
